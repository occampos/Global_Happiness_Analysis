```python
# A happiness score surpassing 6 signifies a state of happiness, while a score below 4.5 indicates unhappiness, with scores falling in-between being categorized as neutral.
# Add signal column

def func(test_data):
    if test_data['happiness_score'] > 6:
        return 'Happy'
    elif test_data['happiness_score'] < 4.5:
        return 'Unhappy' 
    else:
        return 'Moderate'

test_data['signal'] = test_data.apply(func, axis=1)
test_data.head(10)
```
